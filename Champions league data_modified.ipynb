```python
import pandas as pd
import numpy as np
import bar_chart_race as bcr
```

```python
df = pd.read_csv("UEFA Champion League All Season.csv")
df.head()
```

```python
#Remove the Runners-up as what done in the original solution
df.drop(df.loc[df['position']=="runner up"].index, inplace=True)
df
```

```python
#set Season as Index, as what done in the original solution
df = df.set_index("season")
df.head()
```

```python
#if our target is to create an animated bar chart race about the winners, so from my point of view no need to keep ('nation', 'coach', 'formation', 'position')
df.drop(['nation', 'coach', 'formation', 'position'], axis=1, inplace=True)
df.head()
```

### Let's add a new column per each club to can be used later in the animated bar chart

```python
df['club'].value_counts()
```

```python
#here's the list of clubs' names
clubs_list = df['club'].value_counts().index
clubs_list
```

```python
df = pd.concat(
    [
        df,
        pd.DataFrame(
            np.zeros((len(df),len(clubs_list))), 
            columns=clubs_list,
            index=df.index
        )
    ], axis=1
)
df
```

```python
df.info()
```

### Now, it's the time to set values per club/year, so can be used later in the animated bar chart

```python
import warnings
from pandas.core.common import SettingWithCopyWarning
warnings.simplefilter(action="ignore", category=SettingWithCopyWarning)

previousIdx = "-1"
for idx in df.index:
    winner = df['club'][idx]
    for club in clubs_list:
        if club == winner:
            if previousIdx == "-1":
                df[club][idx] = 1
            else:
                df[club][idx] = df[club][previousIdx] + 1
        else:
            if previousIdx != "-1":
                df[club][idx] = df[club][previousIdx]
                
    previousIdx=idx  #to be used in the next iteration
    
df    
```

```python
#now no more need for the club column
del df['club']
df
```

```python
bcr.bar_chart_race(df)
```

```python
bcr.bar_chart_race(df, title = "The Uefa Champions League Winner Since 1956 till last season ",n_bars=5, orientation='h', cmap = 'prism')
```
